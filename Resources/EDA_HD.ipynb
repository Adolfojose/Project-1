```python
# read heart and heart disease prediction csv
heart_data = pd.read_csv(heart_data_to_load)
heart_predict = pd.read_csv(heart_predict_to_load)
```

```python
heart_data_df_cc = heart_data_df.rename(columns={"age": "Age",
                                              "sex": "Sex",                                               
                                              "cp": "Chest pain type",
                                              "trestbps": "BP",   
                                              "chol": "Cholesterol",
                                              "fbs": "FBS over 120",
                                             "exang": "Exercise angina",
                                                "ca": "Number of vessels fluro"})
#disply
heart_data_df_cc
```
